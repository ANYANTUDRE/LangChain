```python
### web based loader
from langchain_community.document_loaders import WebBaseLoader
import bs4

### load, chunk and index the context of the HTML page

loader = WebBaseLoader(web_paths=["https://raamde-bf.net/za%cc%83ndarmiiri-wa%cc%83-yo%cc%83ga-neb-se%cc%83n-da-zuud-sonab%c9%9bl-k%ca%8b%ca%8bra-wa%cc%83/"],
                        bs_kwargs=dict(parse_only=bs4.SoupStrainer(
                            class_=("entry-title", "entry-content")
                        ))
                    )
```

```python
text_docs = loader.load()
text_docs
```

```python
### pdfs load
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("Fondements de la protection de cultures - @CahiersJaunes (1).pdf")
pdf_docs = loader.load()
pdf_docs
```

```python
### splitting into chunks
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
documents = text_splitter.split_documents(pdf_docs)   
documents[:5]
```

```python
### vector mbeddings and vector store
from langchain_community.embeddings import OllamaEmbeddings
from langchain_community.vectorstores import Chroma

db = Chroma.from_documents(documents=documents[:10], embedding=OllamaEmbeddings(model="nomic-embed-text"))
```

```python
# query vector database
query = "Quel est l'impact des ennemis des cultures sur la protection"
result = db.similarity_search(query)
result
```

```python
### FAISS Vector Database

from langchain_community.vectorstores import FAISS
db = FAISS.from_documents(documents=documents[:15], embedding=OllamaEmbeddings(model="nomic-embed-text"))
```

```python
# query vector database
query = "Quel est l'impact des ennemis des cultures sur la protection?"
result = db.similarity_search(query)
result[0].page_content
```

```python
from langchain_community.llms import Ollama

### load mistral model
llm = Ollama(model="mistral")
llm
```

```python
### design chatPrompt template
from langchain_core.prompts import ChatPromptTemplate
prompt = ChatPromptTemplate.from_template(""" Answer the following question based only on the provided context.
                                            <context>
                                                    {context}
                                            </context>
                                            Question:{input}""")
```

```python
### chain intro: create stuff doc chain

from langchain.chains.combine_documents import create_stuff_documents_chain

doc_chain = create_stuff_documents_chain(llm, prompt)
```

```python
### retrievers
retriever = db.as_retriever()
retriever
```

```python
### retrieval chain

from langchain.chains import create_retrieval_chain
retrieval_chain = create_retrieval_chain(retriever, doc_chain)
```

```python
retrieval_chain.invoke({"input": "Quel est l'impact des ennemis des cultures sur la protection?"})
```
